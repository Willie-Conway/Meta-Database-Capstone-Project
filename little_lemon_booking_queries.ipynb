## Task 2: Query the Database to Show All Tables

* Now that you've established the connection, the next step is to query the database and display all the tables.

* Create a query that lists all the tables in the database.

* Execute the query using the cursor object.

* Fetch and print the results.

  Here's the Python code:

```python
# Task 2: Query to show all tables in the database
show_tables_query = "SHOW tables"

# Execute the query to fetch the tables
cursor.execute(show_tables_query)

# Fetch all the results from the executed query
tables = cursor.fetchall()

# Print the tables in the database
print("Tables in the database:")
for table in tables:
    print(table[0])
```

## Task 3: Query with JOIN to Retrieve Customer Data

For this task, you need to join the `Orders` table and the `Customers` table to retrieve the full names and contact details of customers who placed an order greater than $60.

* Identify the required columns from both tables.

* Write the SQL query using the `JOIN` clause and a `WHERE` condition to filter orders with a `TotalCost` greater than 60.

* Execute the query and fetch the results.

  Here's the Python code for Task 3:

```python
# Task 3: Query to retrieve customer details for orders over $60
customer_query = """
SELECT Customers.FirstName, Customers.LastName, Customers.Email, Customers.PhoneNumber, Orders.TotalCost
FROM Customers
INNER JOIN Orders ON Customers.CustomerID = Orders.CustomerID
WHERE Orders.TotalCost > 60;
"""

# Execute the query
cursor.execute(customer_query)

# Fetch the results
customer_results = cursor.fetchall()

# Print the results (full name, contact details, and total cost)
print("Customers with orders over $60:")
for result in customer_results:
    first_name, last_name, email, phone, total_cost = result
    print(f"Name: {first_name} {last_name}, Email: {email}, Phone: {phone}, TotalCost: ${total_cost}")
```
